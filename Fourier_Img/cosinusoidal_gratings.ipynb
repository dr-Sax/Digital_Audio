# 2.3 | Superposition of Two Cosinusoidal Gratings

```python
import numpy as np
import pandas as pd
from scipy.fft import fft2, ifft2, ifftshift, fftshift
import matplotlib.pyplot as plt
%matplotlib inline
```

$$
r_1(x,y) = 1/2 \cos(2 \pi f_1 x) + 1/2
$$


```python
granularity = 0.01
x_max = 5
f1 = 1

x = np.arange(-x_max, x_max, granularity)
X, Y = np.meshgrid(x, x)
r1 = 0.5 * np.cos(2 * np.pi * f1 * X) + 0.5
plt.set_cmap("gray")
plt.imshow(r1)
```

$$
r_2(x,y) = 1/2 \cos(2 \pi f_2[x \cos \theta _2 + y \sin \theta _2]) + 1/2
$$

```python
granularity = 0.01
x_max = 5
f2 = 1
theta2 = -np.pi / 12

x = np.arange(-x_max, x_max, granularity)
X, Y = np.meshgrid(x, x)
r2 = 0.5 * np.cos(2 * np.pi * f2 * (X * np.cos(theta2) + Y * np.sin(theta2))) + 0.5
plt.set_cmap("gray")
plt.imshow(r2)
```

```python
plt.imshow(r1 * r2)
```
